```python
# Medal Count by Country
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Country', data=df, palette='viridis')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Medal Count by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Medal Count by Gender
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Gender', data=df, palette='magma')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Medal Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Medal Count by Sport
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Sport', data=df, palette='rocket')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Medal Count by Sport')
plt.xlabel('Sport')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
